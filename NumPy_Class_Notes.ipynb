**Numpy - Numerical Python**

* `numpy` is a python lib, `written in c-lang` stands for `nummerical python`
* `numpy lib` is vastly used for `mathemthical computations`
* advance usecases of `numpy lib` are `tasks related to images, text data`
* `numpy` is vastly used for data modelling & machine learning tasks as it very efficient in memory mang and faster in execution(`vectorization`)


<b>Numpy Arrays: </b>

* numpy arrays are `homogenous containers`
* `numpy arrays` have `contineous memoy allocation`
* unlike `lists`, `numpy arrays` can perform `elementwise operations\executions`
* in comparrision to efficiency in terms execution speed, `numpy arrays` are close to `50% faster than lists`, and they also `occupy less memory`
* we refer `numpy arrays` as `ndarrays` (`N-Dimenstional Arrays`)

### flatten Vs ravel
* both fallten and ravel are used to convert a higher dimensional ndarray into a 1-d array, however, flatten() method creates a deepcopy of original ndrray where as, ravel() creates a shallowcopy of org ndarray.
* that means any changes to original ndarray will reflect in raveled array and won't reflect in flatten array.
*
