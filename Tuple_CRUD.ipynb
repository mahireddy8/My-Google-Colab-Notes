### <b>Tuple :</b> A tuple is a collection which is ordered and unchangeable or immutable

- it allows duplicates
- it supports indexing
- once tuple created we can not create/ remove elements from tuple
- When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.
- if we create more than one elements within ( ) then only it treat as tuple other it itreats like a string.
- It is also possible to use the tuple() constructor to make a tuple.

- tuple has only 2 methods count() and index()
