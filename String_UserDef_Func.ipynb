#### <b>lower():</b>

```python
def lower():
    s=input("Enter your string here: ")
    l=''
    for i in s:
        if ord(i)>=65 and ord(i)<=90:  # ord('A')--> 65
            l=l+chr(ord(i)+32)         # chr(65) --> A
        else:
            l=l+i
    return l
```

```python
lower()
```

#### <b>upper():</b>

```python
def upper():
    s=input("Enter your string here: ")
    u=''
    for i in s:
        if ord(i)>=65 and ord(i)<=90:
            u=u+i
        else:
            u=u+chr(ord(i)-32)
    return u
```

```python
upper()
```

#### <b>capitalize():</b>

```python
def custom_capitalize(s):
    if 'a' <= s[0] <= 'z':
        first_char = chr(ord(s[0]) - 32)
    else:
        first_char = s[0]
    return first_char + s[1:]
```

```python
custom_capitalize("hey! Python...")
```

####<b> Capitalize():</b>

```python
def cust_capitalize(s):
    if not s:
        return s
    first_char = s[0]
    if 'a' <= first_char <= 'z':
        first_char = chr(ord(first_char) - ord('a') + ord('A'))
    rest = ''
    for char in s[1:]:
        if 'A' <= char <= 'Z':
            rest += chr(ord(char) - ord('A') + ord('a'))
        else:
            rest += char
    return first_char + rest
```

```python
cust_capitalize("hello world")
```

```python
custom_capitalize("mahender")
```

```python
custom_capitalize("vi")
```

```python
def cust_isdigit(s):
    if not s:
        return False
    for char in s:
        if char < '0' or char > '9':
            return False
    return True
```

#### <b>istitle():</b>

```python
def cust_istitle(s):
    if not s:
        return False
    words = s.split()
    for word in words:
        if not ('A' <= word[0] <= 'Z'):
            return False
        for char in word[1:]:
            if 'A' <= char <= 'Z':
                return False
    return True
```

```python
cust_istitle("Hello World")
```

```python
cust_istitle("Apples are in Red")
```

```python
cust_istitle("Few Apples Are In Green")
```

####<b> Title():</b>

```python
def cust_title(s):
    if not s:
        return s
    result = []
    capitalize_next = True
    for char in s:
        if char.isspace():
            result.append(char)
            capitalize_next = True
        elif capitalize_next:
            if 'a' <= char <= 'z':
                result.append(chr(ord(char) - ord('a') + ord('A')))
            else:
                result.append(char)
            capitalize_next = False
        else:
            if 'A' <= char <= 'Z':
                result.append(chr(ord(char) - ord('A') + ord('a')))
            else:
                result.append(char)
    return ''.join(result)
```

```python
cust_title("HELLO MY FRIENDS")
```

```python
cust_title("enter the NeW dragon")
```

```python
cust_title(" ")
```

#### <b>join():</b>

```python
def custom_join(separator, iterable):
    result = ""
    first = True
    for element in iterable:
        if not first:
            result += separator
        result += str(element)
        first = False
    return result
```

```python
custom_join("$",['Hello!','Mahender','reddy'])
```

```python
custom_join("|",['Hello!','Welcome','to','Python'])
```

```python
custom_join("~",['Hi','Welcome','to','Innomtics'])
```
