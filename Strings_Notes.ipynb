```python
# find() and rfind()
s1="apples are in red and apples are good for health some apples in green"
s2='apple'
print("The first occurrence of s2 is at : ", end="")
print(s1.find( s2, 4) )
print("The last occurrence of s2 is at : ", end="")
print(s1.rfind( s2, 4) )
```
