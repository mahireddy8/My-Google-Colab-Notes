#### WAP to check the given number is even or odd ?

```python
n = int(input("Enter your number : "))

if n%2 == 0:
    print("even")
else:
    print("Odd")
```

```python
n = int(input("Enter your number : "))

if n%2 == 1:
    print("ODD")
else:
    print("EVEN")
```

##### WAP to check the given number is +ve or -ve ?

```python
n = int(input("Enter your number : "))
if n>=0:
    print(str(n)+" is Positive")
else:
    print(str(n)+" is negative")
```

### WAP Program check given input Male/FemLe?

```python
inp = input("enter your value : ")
if inp == "m":
    print("Male")
else:
    print("Female")
```

```python
inp = input("enter your value : ")
if inp == "m":
    print("Male")
elif inp == 'f':
    print("female")
else:
    print("Ivalid input")
```

```python
inp = input("enter your value : ")
inp = inp.lower()
if inp == "m":
    print("Male")
elif inp == 'f':
    print("female")
else:
    print("Ivalid input")
```

```python
inp = input("enter your value : ")
if inp == "m" or inp == "M":
    print("Male")
elif inp == 'f' or inp == 'F':
    print("female")
else:
    print("Ivalid input")
```
