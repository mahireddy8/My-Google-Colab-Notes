### Sets

* sets are `hetrogeneous clusters` of `hashableObjects` (`immutable dataObjects`)
* since `sets` they are `clusters` there is `no seq nor order`
* `sets don't entertain duplicate values` (`no duplicates`)
* `sets are mutable`
* we create a set using `set()` or enclosing elements within `{}` seperated by a `,`
* to create an `empty set` we use `set()`. using `{}` will result in an `empty dictionary`

```python
# creating a set
x, y = set([1, 'a', (50, 60)]), {1, 'a', (50, 60)}
print(x, y, sep = '\n')
print(id(x), id(y))
print(type(x), type(y))
```

```python
# create an empty set
a = set()
type(a), len(a)
```

#### difference_update(), intersection_update(), symmetric_difference_update()

* difference_update()	-=	Removes the items in this set that are also included in another, specified set.

* symmetric_difference()	^	Returns a set with the symmetric differences of two sets, Return a set that contains all items from both sets, except items that are present in both sets.

* intersection_update()	&=	Removes the items in this set that are not present in other, specified set(s), Remove the items that is not present in both x and y


```python
fruits = {"apple", "banana", "mango"}
vegetables = {"tomato", "banana", "radish"}

fruits.difference_update(vegetables)

print(fruits)
```

```python
fruits = {"apple", "banana", "mango"}
vegetables = {"tomato", "banana", "radish"}

fv=fruits.symmetric_difference(vegetables)
print(fv)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)
print(x)
```

### Q1. Explore, understand and write notes on :
* Range class and their default values.
* Typecast to a range() and tuple.
-------


In `Python`, the range class represents an immutable sequence of numbers and is commonly used for looping a specific number of times in for loops.

The <b>range</b> `class` has <b>three parameters:</b>

<b>start</b> (optional): The starting value of the sequence. If not provided, it defaults to 0.

<b>stop</b>: The end value of the sequence (exclusive).

<b>step</b> (optional): The step size between each number in the sequence. If not provided, it defaults to 1.

The syntax for using range is:

range([start], stop, [step]).

---------------

```python
for i in range(5):
  print(i)
```

```python
for i in range(0,5):
  print(i)
```

```python
for i in range(0,5,2):
  print(i)
```

```python
for num in range(10, 2, -2): # Negatice step
    print(num) # If you want to decrement, your step must be a negative value.
```
