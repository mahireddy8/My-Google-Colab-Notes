## Strings
* strings are a `seq` of `chars` enclosed in either `single qoutes` or `double qoutes` or `triple single or triple double qoutes`. Strings are `ordered`

* since strings are `seq` and `ordered` we can perform `indexing` and `slicing` of strings

---

* in strings, we can index the chars using either a
  * `positive indexing`
      * (starts qith `0` and ends with `len(stringObject) - 1`)
      `positive indexing` moves from `left to right`

  * `negative indexing`
      * (starts with `-1` and ends with `-len(stringObject)`)
      `negative indexing` moves from `right to left`

> <u>Syntax for indexing</u>

      stringObject[indexValue]

---


* in strings, `slicing` is possible from `either side`. We control this by making use of the `step` value.

> <u>Syntax for slicing</u>

      stringObject[startIndex : endIndex : stepValue]

* default value of stepValue is +1
 * when stepValue is `positive`,
    * default values of `startIndex` and `endIndex` are `0` and `len(stringObject)`
    * slicing happens from `left to right`

  * when the the `stepValue` is 'negative`
    * default `startIndex` and `endIndex` are `-1` and `-(len(stringObject))-1`
    * slicing happens from `right to left`


* `sign` of the `stepValue` indicates the direction.

  * `positive stepValue` indicates `left to right`
  * `negative stepValue` indicates 'right to left`

* `magnitude of stepValue` indicates how many `steps` to taken.

---
* `triple single or triple double qoutes` are used for muliti-line strings, multi-line comments and doc strings.

* `strings` are `immutable` in nature.

#### Placeholders and Modifiers
* A placeholder can contain variables, operations, functions, and modifiers to format the value.
* A placeholder can include a modifier to format the value.
* A modifier is included by adding a colon : followed by a legal formatting type, like .2f which means fixed point number with 2 decimals:



```python
# Adda place holder to the price variable
price = 59
txt = f"The price is {price} dollars"
print(txt)
```

#### Partition()
* The partition() method searches for a specified string, and splits the string into a tuple containing three elements.

* The first element contains the part before the specified string.

* The second element contains the specified string.

* The third element contains the part after the string.

<b>Note:</b> This method searches for the first occurrence of the specified string.

```python
''' Search for the word "bananas", and return a tuple with three elements:

1 - everything before the "match"
2 - the "match"
3 - everything after the "match" '''

txt = "I could eat bananas all day"
x = txt.partition("bananas")
print(x)
```

If the specified value is not found, the partition() method returns a tuple containing: 1 - the whole string, 2 - an empty string, 3 - an empty string:

```python
txt = "I could eat bananas all day"
x = txt.partition("apples")
print(x)
```

#### rpartition()
* The rpartition() method searches for the last occurrence of a specified string, and splits the string into a tuple containing three elements.

* The first element contains the part before the specified string.

* The second element contains the specified string.

* The third element contains the part after the string.

```python
'''
Search for the last occurrence of the word "bananas", and return a tuple with three elements:

1 - everything before the "match"
2 - the "match"
3 - everything after the "match"
'''

txt = "I could eat bananas all day, bananas are my favorite fruit"
x = txt.rpartition("bananas")
print(x)
```

* The zfill() method adds zeros (0) at the beginning of the string, until it reaches the specified length.

* If the value of the len parameter is less than the length of the string, no filling is done.


```python
# Fill the strings with zeros until they are 10 characters long:

a = "hello"
b = "welcome to the Hyderbad"
c = "10.000"

print(a.zfill(10))
print(b.zfill(10))
print(c.zfill(10))
```
