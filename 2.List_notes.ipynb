### Lists
* lists are hetrogeneous, seq and ordered collection datatype.
* lists are mutable
* lists are dynamic
* indexing & slicing are possible on a list.
* a list is created by enclosing elements seperated by, within [ ] or we can also make use of list( )
* an empty list is created by [ ] or list( )


```python
stationary_items=['pen','book','pencil','eraser','sharpender']
print(stationary_items)
```

```python
l1=[1,'a',1.45,3+6j,print,bool]
print(l1,type(l1),id(l1),len(l1))
```

```python
l2=list("hello")
print(l2,type(l2))
```

```python
# empty list
x, y = [], list()
print(x,y)
print(type(x), type(y))
print(len(x), len(y))
```

```python
# creating a list

x = [1, 2.2, 3+3j, 'four', print, str, True, None]
print(x, type(x), len(x), hex(id(x)), sep = '\n')
```

```python
y = list( (1, 2.2, 3+3j, 'four', print, str, True, None) )
print(y, type(y), len(y), hex(id(y)), sep = '\n')
```

### indexing and slicing

```python
# indexing
print(x)
```

```python
x[0], x[-1]
```

```python
# slicing
# left to right
print(x)
print(x[ : : ])
print(x[ : : 1])
print(x[0 : len(x) : 1])
```

```python
# left to right
print(x[ : : -1])
print(x[-1 : -len(x)-1 : -1])
```

```python
print(x[0:7:2])
print(x[-2::-2])
```

### nested lists
* a list is within a list is called as a nested list

```python
z = [1, 2, 3, ['a', 'b', 'c', ['x','y','z']]]
print(z, type(z), len(z), id(z), sep = '\n')
```

```python
z[3][3][1]
```

```python
# ['a', ['x', 'y', 'z']]

print(z[3])
print(z[3][::3])
```
