```python
# calculater program
n1=int(input("Enter a first num: "))
n2=int(input("Enter a second num: "))
operator=input("Enter any operator [+,- ,/,*]")
if operator=='+':
  print(n1,'+',n2,'=',n1+n2)
elif operator=='-':
  print(n1,'-',n2,'=',n1-n2)
elif operator=='*':
  print(n1,'*',n2,'=',n1*n2)
else:
  print('invalid operator')
```

```python
# loginn validation
database={'mahe':2024,'ran':2023,'abhilash':2022}
user=input("Enter a name: ")
if user in database.keys():
  print("User exist")
  password=int(input("Enter a password: "))
  if password==database[user]:
    print("Welcome",user)
  else:
    print('Invalid password')
else:
  print("User not exist")
```
