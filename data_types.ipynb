# Arithmetic Assignment operators

Assignment operators are used to assign values to variables. Arithmetic operators are used to perform mathematical operations on numerical values.

In the example above, we are using the assignment operator += to add 5 to the variable x. This is equivalent to writing x = x + 5.
We can use any of the arithmetic operators in combination with the assignment operator. For example, we could write x -= 2 to subtract 2 from x, or x *= 3 to multiply x by 3.

Assignment operators are a convenient way to write code that performs mathematical operations on variables. They can make your code more concise and easier to read.

# Python Bitwise Operators
<b> Operators </b> are used to perform operations on values and variables. These are the special symbols that carry out arithmetic and logical computations. The value the operator operates on is known as the Operand.

<b>Python Bitwise Operators </b>

Python bitwise operators are used to perform bitwise calculations on integers. The integers are first converted into binary and then operations are performed on each bit or corresponding pair of bits, hence the name bitwise operators. The result is then returned in decimal format.
* Bitwise AND          <b> -->    &  </b>
* Bitwise OR          <b> -->     | </b>
* Bitwise NOT          <b>-->   ~ </b>
* Bitwise XOR          <b>--> ^ </b>
* Bitwise right shift  <b>--> >> </b>
* Bitwise left shift   <b>--> << </b>
